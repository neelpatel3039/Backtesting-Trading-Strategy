## <span style="color: green;"><b><u>Pull Data From Yahoo Finance</u></b></span>

```python
def get_stock_data(ticker, start_date, end_date, interval):
    """
    Retrieves historical stock data for a given ticker symbol within a specified date range and interval.
    
    Parameters:
    ticker (str): The stock ticker symbol (e.g., 'AAPL' for Apple).
    start_date (str): The start date for the data retrieval (format: 'YYYY-MM-DD').
    end_date (str): The end date for the data retrieval (format: 'YYYY-MM-DD').
    interval (str): The data interval (e.g., '1d' for daily data, '1wk' for weekly data).
    
    Returns:
    DataFrame: A pandas DataFrame containing the historical stock data with columns: 'Open', 'High', 'Low', 'Close', 'Volume'.
    """
    
    # Download historical data for the given ticker
    df = yf.download(ticker, start=start_date, end=end_date, interval=interval)
    df.dropna(inplace=True)  # Remove any rows with missing values
    
    # Drop the 'Adj Close' column if it exists
    if 'Adj Close' in df.columns:
        df.drop(columns=['Adj Close'], inplace=True)
    
    # Ensure single-level column names
    df.columns = df.columns.get_level_values(0)  # Flatten multi-level columns, if any
    df.columns.name = None  # Remove the name attribute from columns
    
    # Reset index to include Date column
    df.reset_index(inplace=True)
    df.set_index('Date', inplace=True)  # Set 'Date' column as the index
    
    # Ensure columns are in the correct order
    df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
    
    return df
```

```python
def beautiful_print(df):
    print(df.to_markdown(tablefmt="grid"))
```

```python
# Function to calculate and print fund metrics
def calculate_metrics(trades_df, df, start_date, end_date, initial_fund, display_output=True):
    # Check if trades_df is empty
    if trades_df.empty:
        return {
            "Start Value": initial_fund,
            "End Value": initial_fund,
            "Return %": 0,
            "Buy and Hold Return %": 0
        }

    start_value = initial_fund
    end_value = trades_df['Fund Status'].iloc[-1]
    return_pct = ((end_value - start_value) / start_value) * 100
    start_price = df['Close'].iloc[0]
    end_price = df['Close'].iloc[-1]
    initial_quantity = initial_fund // start_price
    
    # Calculate starting and ending values for buy and hold
    starting_value = start_price * initial_quantity
    ending_value = end_price * initial_quantity
    buy_and_hold_return = ((ending_value - starting_value) / starting_value) * 100

    if display_output:
        # Print approach summaries
        print("-" * 30)
        print("Approach 1 - Buy and Hold")
        print("-" * 30)
        print(f"Fund Start Value: ${start_value:.2f}")
        print(f"Starting Value: ${starting_value:.2f}")
        print(f"Ending Value: ${ending_value:.2f}")
        print(f"Buy and Hold Return %: {buy_and_hold_return:.2f}%")
        print("-" * 30)
        print("\n")
        print("-" * 30)
        print("Approach 2 - Trade and Earn")
        print("-" * 30)
        print(f"Fund Start Value: ${start_value:.2f}")
        print(f"Fund Final Value: ${end_value:.2f}")
        print(f"Return %: {return_pct:.2f}%")
        print("-" * 30)
        print("\n")
        print("-" * 30)
        print("Comparison Chart of Buying and Holding Vs Trading and Earning")
        print("-" * 30)
        
        # Plot comparison bar chart
        fig, ax = plt.subplots()
        approaches = ['Buy and Hold', 'Trade and Earn']
        returns = [buy_and_hold_return, return_pct]
        ax.bar(approaches, returns, color=['blue', 'green'])
        ax.set_ylabel('Return %')
        ax.set_title('Comparison of Returns')
        plt.show()
        
        # Print the trades table
        print("\n")
        print("-" * 30)
        print("Trades Table:")
        print("-" * 30)
        beautiful_print(trades_df)

    return {
        "Start Value": start_value,
        "End Value": end_value,
        "Return %": return_pct,
        "Buy and Hold Return %": buy_and_hold_return
    }
```

### <b> 1. </b> Moving Average Ribbon Strategy

```python
class MovingAverageRibbonStrategy(Strategy):
    short_window = 10  # Shortest moving average
    long_window = 50  # Longest moving average
    stop_loss_percent = 0.02  # Stop loss at 2%

    def init(self):
        # Convert Close prices to Pandas Series and calculate moving averages
        close_series = pd.Series(self.data.Close, index=self.data.index)
        # Short moving average
        self.short_ma = self.I(lambda: close_series.rolling(self.short_window).mean())
        # Long moving average
        self.long_ma = self.I(lambda: close_series.rolling(self.long_window).mean())
        
        self.position_status = 0  # 0: No position, 1: Long position
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        # DataFrame to track trades
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])

    def next(self):
        # Ensure there's enough data to calculate the long moving average
        if len(self.data) < self.long_window:
            return

        # If no position is held, check for buy signal
        if self.position_status == 0:
            # Buy when short moving average crosses above the long moving average
            if crossover(self.short_ma, self.long_ma):
                self.buy()
                self.position_status = 1  # Update to long position
                self.buy_date = self.data.index[-1].date()  # Record buy date without time
                self.buy_price = self.data.Close[-1]
                self.quantity = self.cash // self.buy_price  # Calculate the quantity of stocks bought
                self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash

        # If position is held, check for sell signal
        elif self.position_status == 1:
            # Sell when long moving average crosses above the short moving average
            if crossover(self.long_ma, self.short_ma):
                self.sell()
                self.position_status = 0  # Update to no position
                sell_date = self.data.index[-1].date()  # Record sell date without time
                sell_price = self.data.Close[-1]
                pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
                self.cash += self.quantity * sell_price  # Add selling price to cash
                
                # Record the trade in trades DataFrame
                new_trade = pd.DataFrame({
                    'Buy Date': [self.buy_date],
                    'Sell Date': [sell_date],
                    'Buy Prc': [round(self.buy_price, 2)],
                    'Buy Qty': [self.quantity],
                    'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                    'Sell Prc': [round(sell_price, 2)],
                    'Sell Qty': [self.quantity],
                    'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                    'PnL': [round(pnl, 2)],
                    'Fund Status': [round(self.cash, 2)]
                })
                self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### <b> 2. </b> SuperTrend Strategy

```python
def ATR(df, n=14):
    """
    Calculate the Average True Range (ATR)
    
    Parameters:
    df (DataFrame): DataFrame containing 'High', 'Low', and 'Close' columns.
    n (int): Period for calculating the ATR.
    
    Returns:
    DataFrame: DataFrame with the ATR values.
    """
    # Calculate True Range components
    df['H-L'] = abs(df['High'] - df['Low'])
    df['H-PC'] = abs(df['High'] - df['Close'].shift(1))
    df['L-PC'] = abs(df['Low'] - df['Close'].shift(1))
    
    # Calculate True Range
    df['TR'] = df[['H-L', 'H-PC', 'L-PC']].max(axis=1)
    
    # Calculate ATR
    df['ATR'] = df['TR'].rolling(n).mean()
    
    # Drop intermediate columns
    df.drop(['H-L', 'H-PC', 'L-PC', 'TR'], axis=1, inplace=True)
    
    return df

def SuperTrend(df, period=7, multiplier=3):
    """
    Calculate the SuperTrend indicator
    
    Parameters:
    df (DataFrame): DataFrame containing 'High', 'Low', and 'Close' columns.
    period (int): Period for calculating the ATR.
    multiplier (int): Multiplier for the ATR to set the bands.
    
    Returns:
    DataFrame: DataFrame with the SuperTrend values.
    """
    # Calculate ATR for the given period
    df = ATR(df, period)
    
    # Calculate Upper and Lower Bands
    df['Upper Band'] = ((df['High'] + df['Low']) / 2) + (multiplier * df['ATR'])
    df['Lower Band'] = ((df['High'] + df['Low']) / 2) - (multiplier * df['ATR'])
    
    # Initialize SuperTrend column
    df['SuperTrend'] = 0.0

    # Calculate SuperTrend values
    for current in range(1, len(df.index)):
        if df['Close'][current] > df['Upper Band'][current - 1]:
            df['SuperTrend'][current] = df['Lower Band'][current]
        elif df['Close'][current] < df['Lower Band'][current - 1]:
            df['SuperTrend'][current] = df['Upper Band'][current]
        else:
            df['SuperTrend'][current] = df['SuperTrend'][current - 1]

    return df

class SuperTrendStrategy(Strategy):
    """
    SuperTrend Trading Strategy
    
    Implements a trading strategy based on the SuperTrend indicator.
    """
    def init(self):
        """
        Initialize the strategy by calculating the SuperTrend indicator.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Calculate SuperTrend
        supertrend_df = SuperTrend(pd.DataFrame(self.data.df))
        self.supertrend = self.I(lambda: supertrend_df['SuperTrend'])

        self.position_status = 0  # 0: No position, 1: Long position
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        
        # DataFrame to track trades
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])

    def next(self):
        """
        Define the trading logic for each step.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Check if no position is held and Close price is above SuperTrend
        if not self.position_status and self.data.Close[-1] > self.supertrend[-1]:
            self.buy()
            self.position_status = 1  # Update to long position
            self.buy_date = self.data.index[-1].date()  # Record buy date without time
            self.buy_price = self.data.Close[-1]
            self.quantity = self.cash // self.buy_price  # Calculate the quantity of stocks bought
            self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash

        # Check if a position is held and Close price is below SuperTrend
        elif self.position_status and self.data.Close[-1] < self.supertrend[-1]:
            self.sell()
            self.position_status = 0  # Update to no position
            sell_date = self.data.index[-1].date()  # Record sell date without time
            sell_price = self.data.Close[-1]
            pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
            self.cash += self.quantity * sell_price  # Add selling price to cash
            
            # Record the trade in trades DataFrame
            new_trade = pd.DataFrame({
                'Buy Date': [self.buy_date],
                'Sell Date': [sell_date],
                'Buy Prc': [round(self.buy_price, 2)],
                'Buy Qty': [self.quantity],
                'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                'Sell Prc': [round(sell_price, 2)],
                'Sell Qty': [self.quantity],
                'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                'PnL': [round(pnl, 2)],
                'Fund Status': [round(self.cash, 2)]
            })
            self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### <b> 3. </b> Volatility Breakout Strategy

```python
class VolatilityBreakoutStrategy(Strategy):
    atr_period = 14  # Period for Average True Range (ATR)
    breakout_multiplier = 0.01 # Use a reasonable multiplier for breakout level

    def init(self):
        """
        Initialize the strategy by calculating the Average True Range (ATR).
        
        ATR:
            - atr_period: Number of periods for calculating the ATR.
        """
        # Calculate the Average True Range (ATR) to measure volatility
        self.atr = self.I(lambda: talib.ATR(self.data.High, self.data.Low, self.data.Close, self.atr_period))
        
        self.position_status = 0  # 0: No position, 1: Long position
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])

    def next(self):
        """
        Implement the trading logic based on breakout levels using ATR and ensure selling on the last day.
        
        Steps:
        1. Ensure there is enough data to calculate ATR.
        2. Calculate the dynamic breakout level.
        3. Check for buy/sell signals based on the breakout level and current position.
            - Buy signal: Close price crosses above the breakout level.
            - Sell signal: Close price crosses below the breakout level.
        4. Ensure the stock is sold on the last day.
        """
        # Step 1: Ensure there is enough data to calculate ATR
        # print(f"Checking if enough data for ATR: Data Length={len(self.data)}, ATR Period={self.atr_period}")
        if len(self.data) < self.atr_period:
            print("Not enough data to calculate ATR.")
            return

        # Step 2: Calculate the dynamic breakout level
        self.breakout_level = self.data.Close[-1] + self.breakout_multiplier * self.atr[-1]
        # print(f"Breakout Level: {self.breakout_level}, Close Price: {self.data.Close[-1]}, Breakout Multiplier: {self.breakout_multiplier}")

        # Step 3: Check for buy/sell signals based on the breakout level and current position
        if self.position_status == 0:
            # print("Position Status: No position (0)")
            if self.data.Close[-1] >= self.breakout_level:
                # Step 3.1: Buy signal - Close price crosses above the breakout level
                quantity = self.cash // self.data.Close[-1]
                # print(f"Attempting to Buy: Close={self.data.Close[-1]}, Breakout Level={self.breakout_level}, Quantity={quantity}, Cash={self.cash}")
                if quantity >= 1:  # Ensure at least 1 stock is bought
                    self.buy()
                    self.position_status = 1  # Update to long position
                    self.buy_date = self.data.index[-1].date()  # Record buy date without time
                    self.buy_price = self.data.Close[-1]
                    self.quantity = quantity  # Calculate the quantity of stocks bought
                    self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash
                    # print(f"Bought: Date={self.buy_date}, Price={self.buy_price}, Quantity={self.quantity}, Cash Remaining={self.cash}")
        elif self.position_status == 1:
            # print("Position Status: Holding position (1)")
            sell_breakout_level = self.data.Close[-1] - self.breakout_multiplier * self.atr[-1]  # Calculate sell breakout level
            print(f"Sell Breakout Level: {sell_breakout_level}, Close Price: {self.data.Close[-1]}")
            if self.data.Close[-1] < sell_breakout_level:
                # Step 3.2: Sell signal - Close price crosses below the breakout level
                self.sell()
                self.position_status = 0  # Update to no position
                sell_date = self.data.index[-1].date()  # Record sell date without time
                sell_price = self.data.Close[-1]
                pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
                self.cash += self.quantity * sell_price  # Add selling price to cash
                # print(f"Sold: Date={sell_date}, Price={sell_price}, Quantity={self.quantity}, PnL={pnl}, Cash={self.cash}")
                
                new_trade = pd.DataFrame({
                    'Buy Date': [self.buy_date],
                    'Sell Date': [sell_date],
                    'Buy Prc': [round(self.buy_price, 2)],
                    'Buy Qty': [self.quantity],
                    'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                    'Sell Prc': [round(sell_price, 2)],
                    'Sell Qty': [self.quantity],
                    'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                    'PnL': [round(pnl, 2)],
                    'Fund Status': [round(self.cash, 2)]
                })
                self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)

        # Step 4: Ensure the stock is sold on the last day
        if self.data.index[-1] == self.data.index[-1]:
            if self.position_status == 1:
                # Force sell on last day
                self.sell()
                self.position_status = 0  # Update to no position
                sell_date = self.data.index[-1].date()  # Record sell date without time
                sell_price = self.data.Close[-1]
                pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
                self.cash += self.quantity * sell_price  # Add selling price to cash
                print(f"Forced Sold on Last Day: Date={sell_date}, Price={sell_price}, Quantity={self.quantity}, PnL={pnl}, Cash={self.cash}")

                new_trade = pd.DataFrame({
                    'Buy Date': [self.buy_date],
                    'Sell Date': [sell_date],
                    'Buy Prc': [round(self.buy_price, 2)],
                    'Buy Qty': [self.quantity],
                    'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                    'Sell Prc': [round(sell_price, 2)],
                    'Sell Qty': [self.quantity],
                    'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                    'PnL': [round(pnl, 2)],
                    'Fund Status': [round(self.cash, 2)]
                })
                self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### <b> 4. </b> Money Flow Index (MFI) Strategy

```python
class MFI_Strategy(Strategy):
    """
    Money Flow Index (MFI) Trading Strategy
    
    Implements a trading strategy based on the Money Flow Index (MFI).
    """
    window_size = 14  # Number of periods for calculating the MFI

    def init(self):
        """
        Initialize the strategy by calculating the Money Flow Index (MFI).
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Initialize the necessary variables
        self.money_flow = self.data.Volume * (self.data.Close - self.data.Low) - (self.data.High - self.data.Close)
        self.mfi = self.I(self.calculate_mfi)  # Calculate MFI values
        # Print MFI values for debugging
        print(self.mfi)

        # Initialize trade tracking
        self.position_status = 0  # 0: No position, 1: Long position
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        # DataFrame to track trades
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])

    def calculate_mfi(self):
        """
        Compute the Money Flow Index (MFI) values.
        
        Parameters:
        None
        
        Returns:
        np.array: MFI values
        """
        # Compute the MFI values
        mf_ratio = self.money_flow / self.data.Volume
        # Ensure the output has the same length as the input data
        mfi_values = pd.Series(mf_ratio).rolling(window=self.window_size).mean()  # Adjust rolling window size as needed
        # Pad the first few values with NaN or zeros, depending on your requirements
        return mfi_values.fillna(0).values  # Or use .shift() if you need specific lag handling

    def rolling_sum(self, array, period):
        """
        Create a rolling sum of the array with the given period.
        
        Parameters:
        array (np.array): Input array to calculate the rolling sum.
        period (int): Number of periods for the rolling sum.
        
        Returns:
        np.array: Rolling sum values
        """
        return np.convolve(array, np.ones(period), 'valid')

    def next(self):
        """
        Implement the trading logic based on MFI.
        
        Steps:
        1. Ensure the strategy has valid MFI values before proceeding.
        2. Trading logic based on MFI:
            - Buy signal: MFI is oversold (below 20) and no position is held.
            - Sell signal: MFI is overbought (above 80) and a position is held.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Step 1: Ensure the strategy has valid MFI values
        if np.isnan(self.mfi[-1]):
            return

        # Step 2: Trading logic based on MFI
        # Buy signal: MFI is oversold (below 20) and no position is held
        #print(self.mfi[-1])
        #print(self.position_status == 0 and self.mfi[-1] < 0.2)
        if self.position_status == 0 and self.mfi[-1] < 0.2:
            print("Buying")
            self.buy()
            self.position_status = 1  # Update to long position
            self.buy_date = self.data.index[-1].date()  # Record buy date without time
            self.buy_price = self.data.Close[-1]
            self.quantity = self.cash // self.buy_price  # Calculate the quantity of stocks bought
            self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash
            print(self.cash)

        # Sell signal: MFI is overbought (above 80) and a position is held
        elif self.position_status == 1 and self.mfi[-1] > 0.8:
            print("Selling")
            self.sell()
            self.position_status = 0  # Update to no position
            sell_date = self.data.index[-1].date()  # Record sell date without time
            sell_price = self.data.Close[-1]
            pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
            self.cash += self.quantity * sell_price  # Add selling price to cash
            
            # Record the trade in trades DataFrame
            new_trade = pd.DataFrame({
                'Buy Date': [self.buy_date],
                'Sell Date': [sell_date],
                'Buy Prc': [round(self.buy_price, 2)],
                'Buy Qty': [self.quantity],
                'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                'Sell Prc': [round(sell_price, 2)],
                'Sell Qty': [self.quantity],
                'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                'PnL': [round(pnl, 2)],
                'Fund Status': [round(self.cash, 2)]
            })
            self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### <b> 5. </b> Volume Price Trend (VPT) Strategy

```python
class VPT_Strategy(Strategy):
    """
    Volume Price Trend (VPT) Trading Strategy
    
    Implements a trading strategy based on the Volume Price Trend indicator.
    """
    def init(self):
        """
        Initialize the strategy by calculating the VPT indicator.
        
        Parameters:
        None
        
        Returns:
        None
        """
        self.vpt = self.I(self.calculate_vpt)  # Calculate VPT
        
        self.position_status = 0  # 0: No position, 1: Long position
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        
        # DataFrame to track trades
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])

    def calculate_vpt(self):
        """
        Calculate the Volume Price Trend (VPT) indicator.
        
        Parameters:
        None
        
        Returns:
        Series: VPT values
        """
        # Convert Close and Volume prices to Pandas Series
        close = pd.Series(self.data.Close)
        volume = pd.Series(self.data.Volume)
        
        # Calculate price change percentage
        price_change = close.pct_change()
        
        # Calculate VPT
        vpt = ((price_change * volume).cumsum()).fillna(0)
        
        return vpt
    
    def next(self):
        """
        Define the trading logic for each step.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Ensure there are enough VPT values for comparison
        if len(self.vpt) < 2:
            return

        # Buy signal: VPT is increasing and no position is held
        if self.vpt[-1] > self.vpt[-2] and self.position_status == 0:
            self.buy()
            self.position_status = 1  # Update to long position
            self.buy_date = self.data.index[-1].date()  # Record buy date without time
            self.buy_price = self.data.Close[-1]
            self.quantity = self.cash // self.buy_price  # Calculate the quantity of stocks bought
            self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash

        # Sell signal: VPT is decreasing and a position is held
        elif self.vpt[-1] < self.vpt[-2] and self.position_status == 1:
            self.sell()
            self.position_status = 0  # Update to no position
            sell_date = self.data.index[-1].date()  # Record sell date without time
            sell_price = self.data.Close[-1]
            pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
            self.cash += self.quantity * sell_price  # Add selling price to cash
            
            # Record the trade in trades DataFrame
            new_trade = pd.DataFrame({
                'Buy Date': [self.buy_date],
                'Sell Date': [sell_date],
                'Buy Prc': [round(self.buy_price, 2)],
                'Buy Qty': [self.quantity],
                'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                'Sell Prc': [round(sell_price, 2)],
                'Sell Qty': [self.quantity],
                'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                'PnL': [round(pnl, 2)],
                'Fund Status': [round(self.cash, 2)]
            })
            self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### <b> 6. </b> Heikin-Ashi Candlesticks Strategy

```python
class HeikinAshiStrategy(Strategy):
    """
    Heikin-Ashi Trading Strategy
    
    Implements a trading strategy based on Heikin-Ashi candles.
    """
    def init(self):
        """
        Initialize strategy by calculating Heikin-Ashi candles.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Heikin-Ashi calculations
        self.ha_open = self.I(self.calculate_ha_open)
        self.ha_close = self.I(self.calculate_ha_close)
        
        # Track signals
        self.signal = np.zeros(len(self.data.Close))
        self.signal = self.I(self.calculate_signals)

        # Initialize trade tracking
        self.position_status = 0  # 0: No position, 1: Long position
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        
        # DataFrame to track trades
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])
        
    def calculate_ha_close(self):
        """
        Calculate Heikin-Ashi Close as the average of Open, High, Low, and Close.
        
        Parameters:
        None
        
        Returns:
        Series: Heikin-Ashi Close values
        """
        return (self.data.Open + self.data.High + self.data.Low + self.data.Close) / 4

    def calculate_ha_open(self):
        """
        Calculate Heikin-Ashi Open as the average of the previous Open and Close.
        
        Parameters:
        None
        
        Returns:
        Series: Heikin-Ashi Open values
        """
        ha_open = np.zeros_like(self.data.Close)
        ha_open[0] = (self.data.Open[0] + self.data.Close[0]) / 2  # Initialize
        for i in range(1, len(self.data.Close)):
            ha_open[i] = (ha_open[i - 1] + self.calculate_ha_close()[i - 1]) / 2
        return ha_open

    def calculate_signals(self):
        """
        Generate buy/sell signals based on Heikin-Ashi candle patterns.
        Buy when HA_Close > HA_Open and Sell when HA_Close < HA_Open.
        
        Parameters:
        None
        
        Returns:
        np.array: Buy/Sell signals
        """
        ha_open = self.calculate_ha_open()
        ha_close = self.calculate_ha_close()
        
        signal = np.zeros(len(self.data.Close))
        signal[ha_close > ha_open] = 1  # Buy signal
        signal[ha_close < ha_open] = -1  # Sell signal
        return signal

    def next(self):
        """
        Define trading logic:
        - Buy when signal is 1.
        - Sell when signal is -1 and position exists.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Buy signal: HA_Close > HA_Open and no position is held
        if self.position_status == 0 and self.signal[-1] == 1:
            self.buy()
            self.position_status = 1  # Update to long position
            self.buy_date = self.data.index[-1].date()  # Record buy date without time
            self.buy_price = self.data.Close[-1]
            self.quantity = self.cash // self.buy_price  # Calculate the quantity of stocks bought
            self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash

        # Sell signal: HA_Close < HA_Open and a position is held
        elif self.position_status == 1 and self.signal[-1] == -1:
            self.sell()
            self.position_status = 0  # Update to no position
            sell_date = self.data.index[-1].date()  # Record sell date without time
            sell_price = self.data.Close[-1]
            pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
            self.cash += self.quantity * sell_price  # Add selling price to cash
            
            # Record the trade in trades DataFrame
            new_trade = pd.DataFrame({
                'Buy Date': [self.buy_date],
                'Sell Date': [sell_date],
                'Buy Prc': [round(self.buy_price, 2)],
                'Buy Qty': [self.quantity],
                'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                'Sell Prc': [round(sell_price, 2)],
                'Sell Qty': [self.quantity],
                'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                'PnL': [round(pnl, 2)],
                'Fund Status': [round(self.cash, 2)]
            })
            self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### <b> 7. </b> Renko Chart Strategy

```python
def calculate_renko(df, brick_size):
    """
    Calculate Renko bricks for the given DataFrame and brick size.
    
    Parameters:
    df (DataFrame): DataFrame containing 'Close' prices.
    brick_size (int): Size of each Renko brick.
    
    Returns:
    DataFrame: DataFrame with Renko bricks and their direction.
    """
    df['Renko'] = np.nan
    df['Renko_Direction'] = np.nan
    renko_bricks = []
    direction = None

    for i in range(1, len(df)):
        if df['Close'][i] >= df['Close'][i-1] + brick_size:
            direction = 1  # 'Up' -> 1
            renko_bricks.append(df['Close'][i-1] + brick_size)
        elif df['Close'][i] <= df['Close'][i-1] - brick_size:
            direction = -1  # 'Down' -> -1
            renko_bricks.append(df['Close'][i-1] - brick_size)
        else:
            renko_bricks.append(renko_bricks[-1] if renko_bricks else df['Close'][i])
        
        df.loc[df.index[i], 'Renko'] = renko_bricks[-1]
        df.loc[df.index[i], 'Renko_Direction'] = direction

    return df[['Renko', 'Renko_Direction']]

class RenkoStrategy(Strategy):
    """
    Renko Trading Strategy
    
    Implements a trading strategy based on Renko bricks.
    """
    brick_size = 2

    def init(self):
        """
        Initialize the strategy by calculating Renko bricks.
        
        Parameters:
        None
        
        Returns:
        None
        """
        close_prices = self.data.df['Close']
        renko_df = calculate_renko(pd.DataFrame({'Close': close_prices}), brick_size=self.brick_size)
        
        # Store Renko bricks and direction
        self.renko = self.I(lambda: renko_df['Renko'].reindex(close_prices.index).fillna(method='ffill').values)
        self.renko_direction = self.I(lambda: renko_df['Renko_Direction'].reindex(close_prices.index).fillna(method='ffill').values)
        
        # Initialize trade tracking
        self.position_status = 0  # 0: No position, 1: Long position
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        # DataFrame to track trades
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])

    def next(self):
        """
        Define the trading logic for each step.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Step 1: Ensure the Renko bricks are valid before proceeding
        if len(self.renko) < 2 or np.isnan(self.renko[-1]) or np.isnan(self.renko[-2]):
            return
    
        # Step 2: Check current position status and act accordingly
        if self.position_status == 0:
            # Step 2.1: If no position is held and Renko direction is 'Up', initiate a buy
            if self.renko_direction[-1] == 1:
                self.buy()
                self.position_status = 1  # Update to long position
                self.buy_date = self.data.index[-1].date()  # Record buy date without time
                self.buy_price = self.data.Close[-1]
                self.quantity = self.cash // self.buy_price  # Calculate the quantity of stocks bought
                self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash

        elif self.position_status == 1:
            # Step 2.2: If a position is held and Renko direction is 'Down', initiate a sell
            if self.renko_direction[-1] == -1:
                self.sell()
                self.position_status = 0  # Update to no position
                sell_date = self.data.index[-1].date()  # Record sell date without time
                sell_price = self.data.Close[-1]
                pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
                self.cash += self.quantity * sell_price  # Add selling price to cash
                
                # Record the trade in trades DataFrame
                new_trade = pd.DataFrame({
                    'Buy Date': [self.buy_date],
                    'Sell Date': [sell_date],
                    'Buy Prc': [round(self.buy_price, 2)],
                    'Buy Qty': [self.quantity],
                    'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                    'Sell Prc': [round(sell_price, 2)],
                    'Sell Qty': [self.quantity],
                    'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                    'PnL': [round(pnl, 2)],
                    'Fund Status': [round(self.cash, 2)]
                })
                self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### <b> 8. </b> Mean Reversion with Bollinger Bands and RSI Strategy

```python
class MeanReversionBBandRSI(Strategy):
    """
    Mean Reversion Strategy using Bollinger Bands and RSI
    
    Implements a mean reversion trading strategy based on Bollinger Bands and RSI.
    """
    bb_window = 20  # Number of periods for calculating the moving average (Bollinger Bands)
    rsi_window = 14  # Number of periods for calculating the Relative Strength Index (RSI)
    upper_bb = 2  # Standard deviations above the moving average for the upper Bollinger Band
    lower_bb = 2  # Standard deviations below the moving average for the lower Bollinger Band

    def init(self):
        """
        Initialize the strategy by calculating Bollinger Bands and RSI.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Calculate Bollinger Bands
        close_prices = self.data.Close
        self.upper_band, self.middle_band, self.lower_band = self.I(talib.BBANDS, close_prices, timeperiod=self.bb_window, nbdevup=self.upper_bb, nbdevdn=self.lower_bb)
        
        # Calculate RSI
        self.rsi = self.I(talib.RSI, close_prices, timeperiod=self.rsi_window)
        
        # Initialize trade tracking
        self.position_status = 0  # 0: No position, 1: Long position
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        # DataFrame to track trades
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])

    def next(self):
        """
        Implement the trading logic based on Bollinger Bands and RSI.
        
        Steps:
        1. Ensure that both Bollinger Bands and RSI are valid before proceeding.
        2. Prevent selling if no position is held.
        3. Trading logic based on Bollinger Bands and RSI:
            - Buy signal: Close price is below the lower Bollinger Band and RSI is below 30.
            - Sell signal: Close price is above the upper Bollinger Band and RSI is above 70.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Step 1: Ensure that both Bollinger Bands and RSI are valid before proceeding
        if np.isnan(self.upper_band[-1]) or np.isnan(self.lower_band[-1]) or np.isnan(self.rsi[-1]):
            return

        # Step 2: Trading logic based on Bollinger Bands and RSI
        if self.position_status == 0:
            if self.data.Close[-1] < self.lower_band[-1] and self.rsi[-1] < 30:
                # Step 3.1: Buy signal - Close price is below the lower Bollinger Band and RSI is below 30
                quantity = self.cash // self.data.Close[-1]
                if quantity >= 1:  # Ensure at least 1 stock is bought
                    self.buy()
                    self.position_status = 1  # Update to long position
                    self.buy_date = self.data.index[-1].date()  # Record buy date without time
                    self.buy_price = self.data.Close[-1]
                    self.quantity = quantity  # Calculate the quantity of stocks bought
                    self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash

        elif self.position_status == 1:
            if self.data.Close[-1] > self.upper_band[-1] and self.rsi[-1] > 70:
                # Step 3.2: Sell signal - Close price is above the upper Bollinger Band and RSI is above 70
                self.sell()
                self.position_status = 0  # Update to no position
                sell_date = self.data.index[-1].date()  # Record sell date without time
                sell_price = self.data.Close[-1]
                pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
                self.cash += self.quantity * sell_price  # Add selling price to cash
                
                # Record the trade in trades DataFrame
                new_trade = pd.DataFrame({
                    'Buy Date': [self.buy_date],
                    'Sell Date': [sell_date],
                    'Buy Prc': [round(self.buy_price, 2)],
                    'Buy Qty': [self.quantity],
                    'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                    'Sell Prc': [round(sell_price, 2)],
                    'Sell Qty': [self.quantity],
                    'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                    'PnL': [round(pnl, 2)],
                    'Fund Status': [round(self.cash, 2)]
                })
                self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### <b> 9. </b> Dual Moving Average Strategy

```python
class DualMovingAverage(Strategy):
    """
    Dual Moving Average Trading Strategy
    
    Implements a trading strategy based on short-term and long-term moving averages.
    """
    short_window = 50  # Number of periods for the short-term moving average
    long_window = 100  # Number of periods for the long-term moving average

    def init(self):
        """
        Initialize the strategy by calculating short-term and long-term moving averages.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Calculate short-term and long-term moving averages
        close_prices = self.data.Close
        self.short_ma = self.I(talib.SMA, close_prices, timeperiod=self.short_window)
        self.long_ma = self.I(talib.SMA, close_prices, timeperiod=self.long_window)
        
        # Initialize trade tracking
        self.position_status = 0  # 0: No position, 1: Long position
        self.buy_count = 0  # Count of buy trades
        self.sell_count = 0  # Count of sell trades
        self.cash = 100  # Initial fund amount
        self.quantity = 0  # Quantity of stocks bought
        # DataFrame to track trades
        self.trades_df = pd.DataFrame(columns=[
            'Buy Date', 'Sell Date', 'Buy Prc', 'Buy Qty', 'Buy Prc x Qty', 
            'Sell Prc', 'Sell Qty', 'Sell Prc x Qty', 'PnL', 'Fund Status'
        ])

    def next(self):
        """
        Implement the trading logic based on moving averages.
        
        Steps:
        1. Ensure that both short-term and long-term moving averages are valid before proceeding.
        2. Trading logic based on moving averages crossover:
            - Buy signal: Short-term MA crosses above the long-term MA.
            - Sell signal: Long-term MA crosses above the short-term MA.
        
        Parameters:
        None
        
        Returns:
        None
        """
        # Step 1: Ensure that both short-term and long-term moving averages are valid before proceeding
        if np.isnan(self.short_ma[-1]) or np.isnan(self.long_ma[-1]):
            return

        # Step 2: Trading logic based on moving averages crossover
        if self.position_status == 0:
            # Buy signal: Short-term MA crosses above the long-term MA
            if crossover(self.short_ma, self.long_ma):
                self.buy()
                self.position_status = 1  # Update to long position
                self.buy_count += 1  # Increment buy count
                self.buy_date = self.data.index[-1].date()  # Record buy date without time
                self.buy_price = self.data.Close[-1]
                self.quantity = self.cash // self.buy_price  # Calculate the quantity of stocks bought
                self.cash -= self.quantity * self.buy_price  # Deduct buying price from cash

        elif self.position_status == 1:
            # Sell signal: Long-term MA crosses above the short-term MA
            if crossover(self.long_ma, self.short_ma):
                self.sell()
                self.position_status = 0  # Update to no position
                self.sell_count += 1  # Increment sell count
                sell_date = self.data.index[-1].date()  # Record sell date without time
                sell_price = self.data.Close[-1]
                pnl = (sell_price - self.buy_price) * self.quantity  # Calculate profit or loss
                self.cash += self.quantity * sell_price  # Add selling price to cash
                
                # Record the trade in trades DataFrame
                new_trade = pd.DataFrame({
                    'Buy Date': [self.buy_date],
                    'Sell Date': [sell_date],
                    'Buy Prc': [round(self.buy_price, 2)],
                    'Buy Qty': [self.quantity],
                    'Buy Prc x Qty': [round(self.buy_price * self.quantity, 2)],
                    'Sell Prc': [round(sell_price, 2)],
                    'Sell Qty': [self.quantity],
                    'Sell Prc x Qty': [round(sell_price * self.quantity, 2)],
                    'PnL': [round(pnl, 2)],
                    'Fund Status': [round(self.cash, 2)]
                })
                self.trades_df = pd.concat([self.trades_df, new_trade], ignore_index=True)
```

### Define a Function for Backtesting Trading Strategies

```python
def backtest_strategy(strategy, holdings, cash_per_stock, brokerage, start_date, end_date):
    """
    Backtest a given strategy across multiple stocks.
    
    Parameters:
    strategy (Strategy): The trading strategy to backtest.
    holdings (list): List of stock tickers.
    cash_per_stock (float): Cash allocated per stock.
    brokerage (float): Brokerage commission.
    start_date (str): Start date for historical data.
    end_date (str): End date for historical data.
    
    Returns:
    dict: Dictionary containing strategy results.
    """
    strategy_name = strategy.__name__
    print(f"Backtesting strategy: {strategy_name}")
    
    total_start_value = 0
    total_end_value = 0
    yearly_returns_df = pd.DataFrame()
    
    for ticker in holdings:
        # Download historical data for backtesting
        df = get_stock_data(ticker, start_date, end_date, '1d')
        
        # Backtest the strategy
        bt = Backtest(df, strategy, cash=cash_per_stock, commission=brokerage)
        stats = bt.run()

        # Add ticker symbol to the stats
        stock_stats = stats.to_dict()
        stock_stats['Ticker'] = ticker
        del stock_stats['_equity_curve']
        del stock_stats['_trades']
        all_stats.append(stock_stats)
        
        # Extract the equity curve and calculate yearly returns
        equity_curve = stats['_equity_curve']['Equity']
        yearly_returns = equity_curve.resample('Y').last().pct_change().fillna(0) * 100
        
        # Ensure the DataFrame has all years from start_date to end_date
        all_years = pd.date_range(start=start_date, end=end_date, freq='Y')
        yearly_returns = yearly_returns.reindex(all_years, fill_value=0)
        
        yearly_returns_df[ticker] = yearly_returns.values

        # Calculate metrics for the stock
        stock_metrics = calculate_metrics(stats._strategy.trades_df, df, start_date, end_date, initial_value, display_output=False)
        total_start_value += stock_metrics["Start Value"]
        total_end_value += stock_metrics["End Value"]

    # Calculate the average yearly returns across all stocks for the strategy
    average_yearly_returns = yearly_returns_df.mean(axis=1)
    
    # Create a DataFrame with "Date" and "Average Yearly Returns (%)"
    average_yearly_returns_df = pd.DataFrame({
        'Date': yearly_returns_df.index,
        'Average Yearly Returns (%)': average_yearly_returns
    })
    
    # Calculate overall metrics for the strategy
    overall_return_pct = (total_end_value - total_start_value) / total_start_value * 100 if total_start_value else 0
    
    # Store the results in the dictionary
    strategy_results = {
        'Average Yearly Returns': average_yearly_returns_df,
        'Start Value': total_start_value,
        'End Value': total_end_value,
        'Return %': overall_return_pct
    }
    
    return strategy_results
```
